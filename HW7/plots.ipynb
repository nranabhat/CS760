### Variable Gamma

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read your CSV file
file_path = 'CS760-HW7-release-2/CS760-HW7-release 3/Results/out.csv'
df = pd.read_csv(file_path)

# Extract your data
gamma95 = df.iloc[4450:6450, [2, 3]]
gamma90 = df.iloc[6450:8450, [2, 3]]
gamma85 = df.iloc[8450:10450, [2, 3]]
gamma80 = df.iloc[10450:12450, [2, 3]]
gamma99 = df.iloc[12450:14450, [2, 3]]

# Set the window size for smoothing
window_size = 50  # Adjust as needed

# Apply smoothing using rolling average
smooth_gamma95 = gamma95.rolling(window=window_size).mean()
smooth_gamma90 = gamma90.rolling(window=window_size).mean()
smooth_gamma85 = gamma85.rolling(window=window_size).mean()
smooth_gamma80 = gamma80.rolling(window=window_size).mean()
smooth_gamma99 = gamma99.rolling(window=window_size).mean()

# Plotting with different line styles
plt.figure(figsize=(10, 6))
plt.plot(smooth_gamma95.iloc[:, 0], smooth_gamma95.iloc[:, 1], label='$\gamma$ = 0.95')
plt.plot(smooth_gamma90.iloc[:, 0], smooth_gamma90.iloc[:, 1], label='$\gamma$ = 0.90')
plt.plot(smooth_gamma85.iloc[:, 0], smooth_gamma85.iloc[:, 1], label='$\gamma$ = 0.85')
plt.plot(smooth_gamma80.iloc[:, 0], smooth_gamma80.iloc[:, 1], label='$\gamma$ = 0.80')
plt.plot(smooth_gamma99.iloc[:, 0], smooth_gamma99.iloc[:, 1], label='$\gamma$ = 0.99')

# Adding title and labels
plt.title('Episode Reward over Time for Different $\gamma$ (smoothed over window size 50)')
plt.xlabel('Episode')  # Replace with your actual x-axis label
plt.ylabel('Reward')  # Replace with your actual y-axis label

# Adding legend
plt.legend()

# Show plot
plt.show()
```

### Variable $\alpha$

```python
# Read your CSV file
file_path = 'CS760-HW7-release-2/CS760-HW7-release 3/Results/out.csv'
df = pd.read_csv(file_path)

offset = 10000

alpha0005 = df.iloc[4450 + offset:6450 + offset, [2, 3]]
alpha001 = df.iloc[6450 + offset:8450 + offset, [2, 3]]
alpha002 = df.iloc[8450 + offset:10450 + offset, [2, 3]]
alpha005 = df.iloc[10450 + offset:12450 + offset, [2, 3]]
alpha01 = df.iloc[12450 + offset:14450 + offset, [2, 3]]

# Set the window size for smoothing
window_size = 50  # Adjust as needed

# Apply smoothing using rolling average
smooth_alpha0005 = alpha0005.rolling(window=window_size).mean()
smooth_alpha001 = alpha001.rolling(window=window_size).mean()
smooth_alpha002 = alpha002.rolling(window=window_size).mean()
smooth_alpha005 = alpha005.rolling(window=window_size).mean()
smooth_alpha01 = alpha01.rolling(window=window_size).mean()

# Plotting with different line styles
plt.figure(figsize=(10, 6))
plt.plot(smooth_alpha0005.iloc[:, 0], smooth_alpha0005.iloc[:, 1], label='alpha = 0.0005')
plt.plot(smooth_alpha001.iloc[:, 0], smooth_alpha001.iloc[:, 1], label='alpha = 0.001')
plt.plot(smooth_alpha002.iloc[:, 0], smooth_alpha002.iloc[:, 1], label='alpha = 0.002')
plt.plot(smooth_alpha005.iloc[:, 0], smooth_alpha005.iloc[:, 1], label='alpha = 0.005')
plt.plot(smooth_alpha01.iloc[:, 0], smooth_alpha01.iloc[:, 1], label='alpha = 0.01')

# Adding title and labels
plt.title('Episode Reward over Time for Different alpha (smoothed over window size 50)')
plt.xlabel('Episode')  # Replace with your actual x-axis label
plt.ylabel('Reward')  # Replace with your actual y-axis label

# Adding legend
plt.legend()

# Show plot
plt.show()
```
